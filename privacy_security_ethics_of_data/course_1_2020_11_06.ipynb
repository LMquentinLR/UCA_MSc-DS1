## Privacy

A couple of popular definitions:
- The right to be left alone (focus on freedom from intrusion) -- it comes from 1980 from Warren & Brandeis
- Informational self-determination (focus on control)

The definition of privacy is **abstract** and **subjective**. It depends on study discipline, social norms and expectations.

#### European Convention on Human Rights (1953):

The modern concept of privacy emerged as a response to the excesses of totalitarian states in the 1930s and 1940s: to protect citizens from an overbearing/intrusive state.
- <u>Article 8</u> - Right to respect private and family life
    1. Everyone has the right to respect for his private and family life, his home and his correspondence
    2. There shall be no interference by a public authority with the exercise of this right
    
#### Westin's definition (1970): 

The right of the individual to decide what information about himself should be communicated to others and under what circumstances.

#### Privacy and Informatics Law (French, 1978):

Introduces a definition of 'personal data' (any information relating to an individual). It is prohibited to collect and process personal data from racial, ethnic, political, philosophical or religious type, and data related to health or sexual orientation.

It forms the basis of GDPR.

### General Data Protection Regulation (GDPR) 

It is an extension of the **EU Data Protection Directive** (1995). It was introduced in 2018.

It defines and applies to "**personal data**."

#### Any information can be personal data

Any information can fall under personal data regardless of its **nature** (true or inaccurate, objective and subjective -- including opinions and assessments), **content** (not strict to private or family life and could concern an individual's profesionnal life) or **format** (alphabetical, numerical, graphical,m photographical or acoustif, kept on paper, stored in memory on a computer, structured or unstructured, video and voice recording).

#### Any information can relate to a person in 3 conditions

- **content**: facts about that person's identity, characteristics, or behaviour (e.g. medical, criminal, etc.)

- **purpose**: when data are used or are likely to be used with the puirpose to evaluate, trat in a certain way, influence the status or behaviour of an individual (e.g. person carried unauthorized alterations to their house, etc.)

- **result/impact**: when its use is likely to have an impact on a person's rights and interests (e.g. different treatments, information recorded to monitor the productivity of an employee, etc.)

### Identified or Identifiable

- **identified**: person who is known or distinguished from the others in a group

- **identifiable**: person who is not identified yet but identification is possible

- **directly**: reference to a name, in combination with additional information, if the name is not unique

- **indirectly**: unique combinations of indirect identifiers that allow a person to be singled out from others

e.g. online identifiers: IP address, cookies, RFID tags, MAC addresses, etc.

e.g. direct identifiers: Name, address details, email address, etc.

e.g. indirect identifiers: physical, physiological, genetic, mental, economic, cultural, etc.

### Test of "reasonably likelihood" of identification

To check if a person is identifiable, account to be taken to:
- All means "reasonably likely" to be used to identify an individual directly or indirectly (e.g. puyblic registry, reverse directory)
- By any person (not necessary that all the information to identify must be in the hands of one person (e.g. ordinary person or by a particular person)

**Example of accounting factors**: cost/time needed for identification, intended explicit or implied purpose of processing, available tools for identification, risk of organizational dysfunctions, state of the art of technology at the time of processing.
