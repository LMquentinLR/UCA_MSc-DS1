### Community Consensus

**Five strategies** that guide design intent to manipulate users:

1. <u>Nagging:</u> A minor redireciton of expected functionality that may persis over one or more interactions
2. <u>Obstruction:</u> Impeding a task flow, making an interaction more difficult than it needs to be, with intent to dissuade an action
3. <u>Sneaking:</u> Hiding, disguising, or delaying the divulging of relevant information. Often intended to force uninformed decisions
4. <u>Interface interference:</u> Privileging specific actions over others, therby confusing the user or limiting discoverability of improtant actions
5. <u>Forced Action:</u> Requiring users to perform a specific action to access (or continue to access) specific functionality
