The algorithm we have been considering have a fixed alpha (constant or decreasing alpha with regards to $\frac{1}{k}$). The issue relates to the varying curvature of F. We need to have a small alpha when the curve is steep, and a high alpha when the curve is flatter.

The difficulty relates to the varying curvature of the empirical risk $F$.

**We want a small L and a big C**.

$$w_{k+1} = w_k - \alpha_k.H_F^{-1}.\nabla F(w_k)$$
