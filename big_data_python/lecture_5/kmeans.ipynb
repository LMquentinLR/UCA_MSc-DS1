## 0. Library imports

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize']=(10,8)

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering
```

# KMeans Practice

### Making sample and visualization

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)
print(X.shape, y.shape)
```

```python
plt.scatter(X[:,0], X[:,1])
```

### Implementing the KMeans

**make_blob example**

```python
model = KMeans(n_clusters=4)
model.fit(X)
y_kmeans = model.predict(X)
centers = model.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, cmap="viridis")
plt.scatter(centers[:,0], centers[:,1], c="black", s=500, alpha=0.5, marker="o")
plt.show()
```

**make_moons example**

```python
X, y = make_moons(200, noise=0.05, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
model = KMeans(n_clusters=2)
model.fit(X)
y_kmeans = model.predict(X)
centers = model.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=50, cmap="viridis")
plt.scatter(centers[:,0], centers[:,1], c="black", s=500, alpha=0.5, marker="o")
plt.show()
```

In this case KMeans might not be the best choice to perform a clustering.

**Spectral Clustering Example on Moons**

```python
model = SpectralClustering(n_clusters=2, affinity="nearest_neighbors", assign_labels="kmeans")
labels = model.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, s=50, cmap="viridis")
```
