## Confidence Interval
<hr>


We have a certain parameter $\theta$ to estimate and a $n$-number of observations with which to estimate $\theta$ in a parametric framework.

Let $X_1, ..., X_n \sim â„±(\theta)$

A **confidence interval** is an interval $C=[a,b]$, with $a=a(X_1,...,X_n)$ and $b=(X_1,...,X_n)$ both functions of the data.

A confidence interval of level $1-\alpha$ is an interval of such kind such that $\forall \theta \in \mathbb{R}$,  $P_\theta(\theta \in C) \ge 1-\alpha$.

The random quantity $\theta$ isn't a random variable, but $a$ and $b$ are random variables.
<hr>

<u>Example</u>:

Let $\alpha=0.05$, we have $1-\alpha=0.95$. What is the easiest confidence interval that contains the mean of a Gaussian distribution (Let $X_1, ..., X_n \sim N(\mu, \sigma^2)$)?

> $]-\infty, +\infty[$
<hr>

We have an estimator $\hat{\theta_n}$ used to estimate an unknown parameter $\theta$. Let's assume (based on the Central Limit Theorem) that:
> $\hat{\theta_n} \sim N(\theta,\hat{se}(\hat{\theta_n})^2)$

If we define $Z_{\frac{\alpha}{2}}=\Phi^{-1}(1-\frac{\alpha}{2})$, it means that $P(Z \ge \frac{\alpha}{2})=1-\frac{\alpha}{2}$ with $Z\sim N(0,1)$. 

Z is called the **quantile**.

> $C_\alpha = [\quad\hat{\theta_n}-Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad,\quad\hat{\theta_n}+Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad]$

$C_\alpha$ is a confidence interval with level $1-\alpha$.
<hr>

We need to describe this quantity:

> $P(\quad\hat{\theta_n}-Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad\le\quad\theta\le\quad\hat{\theta_n}+Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad)$

We have:

> $P(\quad-Z_{\frac{\alpha}{2}}\quad\le\quad\frac{\theta-\hat{\theta_n}}{\hat{se}(\hat{\theta_n}}\quad\le\quad Z_{\frac{\alpha}{2}}\quad)$
>
> with $\frac{\theta-\hat{\theta_n}}{\hat{se}(\hat{\theta_n}} \sim N(0,1)$ Sometimes, $\frac{\theta-\hat{\theta_n}}{\hat{se}(\hat{\theta_n}}$ is called $nill Z$ or $Z$

From this equality, and by definition of $Z_{\frac{\alpha}{2}}$, $1-\alpha$, we obtain:

> $P(\quad-Z_{\frac{\alpha}{2}}\quad\le\quad\frac{\theta-\hat{\theta_n}}{\hat{se}(\hat{\theta_n}}\quad\le\quad Z_{\frac{\alpha}{2}}\quad) = P(Z\le Z_{\frac{\alpha}{2}}) - P(Z\le-Z_{\frac{\alpha}{2}}) = \Phi(Z_{\frac{\alpha}{2}}) - \Phi(-Z_{\frac{\alpha}{2}})$ 
>
> with $\Phi(x)=\int_{-\infty}^x \frac{1}{\sqrt{2.\pi}}.e^{\frac{u^2}{2}}du$
> 
> We note: $\Phi(Z_{\frac{\alpha}{2}}) - \Phi(-Z_{\frac{\alpha}{2}}) = 1-\frac{\alpha}{2} - (1-(1-\frac{\alpha}{2})) = 1-\alpha$

## Nonparametric Estimation
Covers Chapter 7 of <u>*All of Statistics, a Concise Course in Statistical Inference*</u>
<hr>

Let $X_1, ..., X_n$ Independent and Identically Distributed (IID) random variables (RV). 

We say: $F(x) = P(X_1\le x)$

We call $\hat{F_n}(x)$ the **empirical distribution function** such that $\hat{F_n}(x)=\frac{1}{n}\underset{i=1}{\overset{n}{\sum}}\mathbb{1}_{]-\infty,x]}(X_i)$ 
<hr>

<u>Exercise 1:</u> Prove $\mathbb{E}[\hat{F_n}(x)]=F(x)$ is an unbiased estimator.

We have:

$\frac{1}{n}\underset{i=1}{\overset{n}{\sum}}\mathbb{E}\mathbb{1}_{]-\infty,x]}(X_i)$

Recall: $\mathbb{E}(h(x)) = \int_{-\infty}^{+\infty}h(u)f(u)du$

We have:

$\mathbb{E}[\hat{F_n}(x)]=\int_{-\infty}^{+\infty}\mathbb{1}_{]-\infty,x]}(u)f(u)du$ with $f$ the probability density function (unknown).

$\mathbb{E}[\hat{F_n}(x)]=\int_{-\infty}^{x}f(u)du = F(x)$ We have our proof.

<hr>

<u>Exercise 2:</u> Prove the proposition: If $\mathbb{E}_\theta[\hat{\theta_n}] = \theta$, we have the condition $\mathbb{V}(\hat{\theta_n}) \underset{n\rightarrow+\infty}{\rightarrow}0 \Rightarrow \hat{\theta_n}\overset{P}{\rightarrow}\theta$

$\mathbb{V}(\hat{F_n}(x))=\frac{1}{n^2}\underset{i=1}{\overset{n}{\sum}}Var(\mathbb{1}_{]-\infty,x]}(X_i))$

$\mathbb{V}(\mathbb{1}_{[-\infty,x]}(X_i))=\mathbb{E}(\mathbb{1}^2_{]-\infty,x]}(X_i))-(\mathbb{E}(\mathbb{1}_{]-\infty,x]}(X_i)))^2$

$\mathbb{E}(\mathbb{1}^2_{[-\infty,x]}(X_i))-(\mathbb{E}(\mathbb{1}_{]-\infty,x]}(X_i)))^2=F(x)-F^2(x)=F(x).(1-F(x))$

Thus:

$\mathbb{V}(\hat{F_n}(x))=\frac{F(x).(1-F(x))}{n}$

We see that $\underset{n\rightarrow+\infty}{lim}\mathbb{V}(\hat{F_n}(x))=0$ and we conclude: 

> $\hat{F_n}(x)\overset{P}{\rightarrow}F(x)$
<hr>

With: $\hat{F_n}(x) = \frac{1}{n}\underset{i=1}{\overset{n}{\sum}}\mathbb{1}_{[-\infty,x]}(X_i)$ We have:

$\mathbb{E}[\hat{F_n}(x)] = F(x)$

$\mathbb{V}[\hat{F_n}(x)] = \frac{F(x).(1-F(x))}{n}$

**Remark**:

$X_1, ..., X_n$ are IID RV. by assumption. Thus $\mathbb{1}_{[-\infty,x]}(X_1)$, ..., $\mathbb{1}_{[-\infty,x]}(X_n)$ are also IID RV.

$\mathbb{V}[\mathbb{1}_{]-\infty,x]}(X_1)]=\frac{F(x).(1-F(x))}{n}$

Using the Central Limit Theorem $\Rightarrow \hat{F_n}(x)\overset{d}{\rightarrow}{N(F(x),\frac{F(x).(1-F(x))}{n})}$
<hr>

We recall: $C_\alpha = [\quad\hat{\theta_n}-Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad,\quad\hat{\theta_n}+Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad]$ a $1-\alpha$ confidence interval.

$C_\alpha = [\quad\hat{\theta_n}-Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad,\quad\hat{\theta_n}+Z_{\frac{\alpha}{2}}.\hat{se}(\hat{\theta_n})\quad]$

We have $\hat{\theta_n}=\hat{F_n}(x)$ and $\hat{se}(\hat{\theta_n})=\sqrt{\frac{F(x).(1-F(x))}{n}}$. However, **we do not know F(x)**. In practice, we replace what we don't know by the estimate:

$\hat{se}(\hat{\theta_n})=\sqrt{\frac{\hat{F_n}(x).(1-\hat{F_n}(x))}{n}}$

$\sqrt{n}.\frac{\hat{\theta_n}-F(x)}{\sqrt{F(x).(1-F(x))}}\overset{d}{\rightarrow} N(0,1)$ and $\frac{\sqrt{F(x).(1-F(x))}}{\sqrt{\hat{F_n}(x).(1-\hat{F_n}(x))}}\overset{P}{\rightarrow}1$

We have:

$\sqrt{n}.\frac{\hat{\theta_n}-F(x)}{\sqrt{\hat{F_n}(x).(1-\hat{F_n}(x))}}.\frac{\sqrt{F(x).(1-F(x))}}{\sqrt{F(x).(1-F(x))}}$

thus:

$\sqrt{n}.\frac{\hat{\theta_n}-F(x)}{\sqrt{\hat{F_n}(x).(1-\hat{F_n}(x))}}.\frac{\sqrt{F(x).(1-F(x))}}{\sqrt{F(x).(1-F(x))}}\underset{n\rightarrow+\infty}{\overset{d}{\rightarrow}}N(0,1).1=N(0,1)$ (Slutskyâ€™s Theorem)
