```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Logistic regression

```python
def logit(p):
    """
    Activation function of the logistic regression
    Perfect for linearly separable items / can be extended to multiclass
    """
    return np.log(p/(1-p))

# sigmoid(logit(p)) = p

def sigmoid(p):
    """
    Takes real values and transform them in the 0,1 range with an intercept at 0.5
    """
    return 1/(1+math.exp(-p))
```

```python
arr = np.linspace(0.01,0.99,100)
```

```python
plt.plot(arr, arr)
plt.show()
```

```python
y = [logit(p) for p in arr]
```

```python
plt.plot(arr, y)
plt.show()
```

```python
y_2 = [sigmoid(p) for p in y]
```

```python
plt.plot(arr, y_2)
plt.show()
```
