## 1. Notions and Definitions

### 1.1 Random Variable (abbrev. rv.)

Random variables define distributions, which can either be discrete or continuous.

1. **discrete**, the distribution has a finite number of variable or can be infinite if `∃ bijection between the set of real values ℝ and the set of integers ℕ`.
2. **continous**, the distribution has a set of possible values that can be written as an interval or union of intervals.

/!\ *Sometimes, discrete variables can be considered as continuous

### 1.2 Distribution

A distribution is defined by two conditions:

> $X(\Omega)$ as the set of possible values for rv. X

> if X is **discrete**: $\forall k \in \mathbb{R}$, parametrized by probabilities $P(X=k)$
>
> if X is **continuous**: $\forall k \in \mathbb{R}$, parametrized by a density function $f_x(x)$

### 1.3 Probability Density Function

A density function is a function such that 2 conditions are met for X a random variable.

> $f_x : \mathbb{R} \to \mathbb{R}$ (i.e. application) such that $\forall k \in \mathbb{R}$, $f_x(x) \ge 0$

*Negative probabilities are forbidden by definition*.


> $\int_{-\infty}^{+\infty}f_x(x)dx = 1$

<u>Remarks:</u>

1. In the **discrete case**: $P(X)>=0$ and $\sum_{\forall k \in X(\Omega)}P(X=k)=1$
2. In the **continuous case**: $P(X=a) = 0$ <p>
    $P(a \le x \le b) = \int_{a}^{b}f_x(x)dx = P(a \lt x \le b) = P(a \le x \lt b) = P(a \lt x \lt b)$ (*This is not true in the case of discrete variables*)

### 1.4 Expectation, Variance, and Moments

$L^P$ is the set of states where all the rv. X fulfills the following condition: $\int_{\mathbb{R}}\vert x \vert^P.f_x(x)dx \lt  +\infty$.

if $X \in L^P \Rightarrow X$ has a **moment** or **order P** and as such:

> $E[X^P] = \int_{\mathbb{R}} x^P.f_x(x)dx$

#### Expectation and Variance, two special cases

1. The Expectation is the moment of order 1 of rv. X
2. The Variance is the moment of order 2 of rv. X

If $X \in L^1$, we defined the Expectation of X by $E[X] = \int_{\mathbb{R}} x . f_x(x)dx$

If $X \in L^2$, we defined the Variance of X by $V[X] = E[X^2] - (E[X])^2 = E[(X-E[X])^2]$

By property, $V \ge 0$.

<u>Remarks:</u>

By property, $P \le q \Rightarrow L^q \subset L^p$. In particular $X \in L^2 \Rightarrow X \in L^1$

### 1.5 (cumulative) Distribution Function

Let X be a rv., the cumulative distribution function $F_x$ for X is given by:

> $\forall t \in \mathbb{R}, F_x(t) = P(X \le t)$
>
> $f_x(t) = F_x^{'}(t)$

A distribution function is a way to characterize a rv.

<u>Remarks:</u>

The above definition is the traditional French one. In English the symbol $\le$ may be replaced with $\lt$.

#### Properties

> $\forall t \in \mathbb{R}, 0 \ge F_x(t) \ge 1$, $t \to F(t)$ is an increasing function

> if X is **discrete**: $F_x(t)$ is a piecewise constant function
>
> if X is **continuous**: $F_x(t)$ is **strictly** increasing on the **support of X**

Support of X: $X(\Omega)=\int_{t\in\mathbb{R}} f_x(t) \gt 0$

if we have two rv. X and Y, then $F_x(t) = F_y(t)$ in $\mathbb{R}$ means that X and Y have the same distribution **but it does not imply that X = Y**

#### Continuity of $F_x$

> if X is **discrete**: $F_x(t)$ is right continuous (in the English notation seen above, it is left continuous)
>
> if X is **continuous**: $F_x(t)$ is continuous

#### Limits of $F_x$

$\lim_{t \to -\infty} F_x(t) = 0$

$\lim_{t \to +\infty} F_x(t) = 1$

## 2. Classical Random Variables

### 2.1 Discrete Case: Bernouilli

Bernouilli rv. depend on 1 parameter $p$ (but 2 rvs.).

> $X(\Omega) = \{0; 1\}$
>
> $0 \le p \le 1$
>
> $P(X=1) = p$; $P(X=0)=1-p$, with $(1-p)$ given its own term $q$

$p$ is traditionally called `probability of success`.

#### Expectation, Variance

> $E[X] = p$
>
> $V[X] = p . q$

#### Moment

Bernouilli rv. have infinite moments because Bernouilli has a **finite**, **discrete** number of rv.

### 2.2 Discrete Case: Binomial

Binomial rv. depend on 2 parameters $n$ and $p$ (but 2 rvs.). Denoted  $X \sim B(n,p)$

> $X(\Omega) = \{0; 1; ...; n\}$
>
> $p \in [0;1]$
>
> $n \in \mathbb{N} \ast$
>
> $\forall k \in X(\Omega), P(X=k) = \binom{n}{k}*p^k.q^{n-k}$

$\binom{n}{k}=\frac{n!}{k!.(n-k)!}$

#### Expectation, Variance

> $E[X] = n.p$
>
> $V[X] = n . p . q$

#### Relationship with Bernouilli

We consider a Bernouilli experiment repeated n times **independently**. We Get $X_1, X_2, ..., X_n$ as n independent rv. with parameter $p$. 

> $X = \sum_{t=1}^nX_t$

X is the number of successes among the collection $X_1, X_2, ..., X_n$.

This modelization is useful for the computation of E[X] and V[X].

#### Properties of E and V

Let X and Y two rv.

Let 𝜆 and 𝜇 two reals.

> $E[𝜆.X+𝜇.Y] = 𝜆.E[X]+𝜇.E[Y]$
>
> $E[𝜇] = 𝜇$

The expectation is **linear**. The variance is not.

> generally $V[X+Y] \neq V[X] + V[Y]$. It is only true when X and Y are independent
>
> $V[𝜆.X] = 𝜆^2.V[X]$
>
> $V[𝜇] = 0$

The variance is **quadratic**.

**Example**. For $X \sim B(n,p)$:

$E[X] = E[X_1+X_2+ ... + X_n]$ with $\forall k \in {1,2,...,n}$, $X_k$ independent and following the Bernouilli distribution.

By linearity:

$E[X] = E[X_1]+E[X_2]+ ... + E[X_n] = p + p + ...+ p = n.p$

For the variance:

$V[X] = V[X_1+X_2+ ... + X_n]$

Because $\forall k \in {1,2,...,n}$, $X_k$ independent and following the Bernouilli distribution:

$V[X] = V[X_1]+V[X_2]+ ... + V[X_n] = q.p + q.p + ...+ q.p = n.q.p$

### 3.3 Discrete Case: Poisson

Poisson rv. depend on 1 parameter 𝜆 with $𝜆 \in \mathbb{R}\ast$

> $X(\Omega) = \mathbb{N}$. It can be zero
>
> $\forall k \in \mathbb{N}, P(X=k) = e^{-𝜆}.\frac{𝜆^k}{k!}$

A stochastic poisson process is a counting process.

#### Expectation, Variance

> $E[X] = 𝜆$
>
> $V[X] = 𝜆$

<u>Teacher's note:</u> A distribution called Hawkes distribution is an improvement on the Poisson distribution.

### 3.4 Continuous Case: Uniform

Denoted  $X \sim U([a,b])$. It is uniform on $[a,b]$

> $X(\Omega) = [a,b]$. It can be zero
>
> $f_x(t) = \frac{1}{b-a}$ if $t \in [a,b]$
> otherwise $f_x(t) = 0$ 

#### Expectation, Variance

> $E[X] = \frac{(a+b)}{2}$
>
> $V[X] = \frac{(b-a)^2}{12}$


$E[X] = \int_a^b t . f_x(t)dt$
$E[X] = \frac{1}{b-a} . \int_a^b t dt$
$E[X] = \frac{1}{b-a} . [\frac{t^2}{2}]_a^b$

```python
print("Summary of the Uniform distribution -- English denotation")
print("Source: https://en.wikipedia.org/wiki/Uniform_distribution_(continuous)")
Image("pictures/en_uniform_distribution.png")
```

### 3.5 Continuous Case: Exponential

Denoted  $X \sim  𝓔(𝜆)$. It is an exponential rv. with 1 param 𝜆.

> $𝜆 \in \mathbb{R}^{+\ast}$
>
> density function: $\forall t \in \mathbb{R}^{\ast}, f_x(t) = e^{-𝜆.t}.𝜆$
> otherwise $f_x(t) = 0$ 

#### Expectation, Variance

> $E[X] = \frac{1}{𝜆}$
>
> $V[X] = \frac{1}{𝜆^2}$


$E[X] = \int_a^b t . f_x(t)dt$
$E[X] = \frac{1}{b-a} . \int_a^b t dt$
$E[X] = \frac{1}{b-a} . [\frac{t^2}{2}]_a^b$
