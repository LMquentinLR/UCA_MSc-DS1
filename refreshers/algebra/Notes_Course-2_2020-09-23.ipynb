## Review of Scalar Products

1. $<v,w>$ is any **bilinear** and **symmetric** map from $VxV\rightarrow\mathbb{R}$. In the context of this course: **positive definite dot product**.
<p>
2. $\forall v \ne 0, <v,v > 0$.
<p>
3. $<v,w>$ will most of the time be of the shape $\forall v, w \in \mathbb{R}^\mathbb{N}, \mathbb{R}^\mathbb{N}.\mathbb{R}^\mathbb{N} = v^T.w = w^T.v = \sum_{i=1}^{N}v_i.w_i$.
<p>
4. $||v|| = \sqrt{<v,v>}$ also called **norm**.
<p>
5. In V, $dim(V) = N, \exists v_1, ..., v_n$ vectors such that $\forall i, j \in \mathbb{R}$ with $v_i \perp v_j$ ($\Rightarrow <v_i,v_j> = 0$) $\Rightarrow \{v_1,...,v_n\}$ forms a basis for V.
<p>
6. if $v \in V$, $v = \sum_{i=1}^N{\frac{<v,v_i>}{||v_i||}.v_i}$ if $\{v_1,...,v_n\}$ is orthonormal (i.e. $||v_i||=1$).

**Definition**: Given a vector space V and two vectors $v, w \in V$, the angle between them is the real number $\theta \in [0,\pi]$ defined by

> $cos\theta := \frac{<v,w>}{||v||.||w||}$

## Proof of Obtaining an Orthogonal Basis from Any Basis

Using the case of $\mathbb{R}^2$ and the **Gram_Schmidt procedure** we consider the orthogonal projection of $v_2$ on $v_1$ denoted $P_{v_1}(V_2)$ (See course pdf for illustration). This projection is aligned with v1 such that:

> $P_{v_1}(V_2) = \alpha * v_1$, $\exists \alpha \in \mathbb{R}^2$

and:

> $w_2:= v_2 - _{v_1}(V_2)$ the vector which, in $\{v_1, w_2\}$ forms an orthogonal basis in $\mathbb{R}^2$

By trigonometry we know that:

> $||P_{v_1}(V_2)|| = \alpha * ||v_1|| = ||v2|| * cos \theta$
>
> $alpha * ||v_1|| = ||v2|| * cos \theta$
>
> $\Rightarrow \alpha = \frac{||v_2||}{||v_1||}*cos\theta$
>
> $\alpha = \frac{||v_2||}{||v_1||}*\frac{<v_1,v_2>}{||v_1||.||v_2||}$
>
> $\alpha = \frac{<v_1,v_2>}{||v_1||^2}$

Thus:

> $P_{v_1}(V_2) = \alpha * v_1 = \frac{<v_1,v_2>}{||v_1||^2} * v_1$

**Proof that $w_2$ is orthogonal to $v_1$**:

> $<v_1,w_2> = <v_1,v_2 - P_{v_1}(V_2)> = <v_1,v_2> - <v_1,P_{v_1}(V_2)>$
>
> $<v_1,w_2> = <v_1,v_2> - <v_1,\frac{<v_1,v_2>}{||v_1||^2} * v_1>$
>
> $<v_1,w_2> = <v_1,v_2> - \frac{<v_1,v_2>}{||v_1||^2} * <v_1,v_1>$
>
> $<v_1,w_2> = <v_1,v_2> - \frac{<v_1,v_2>}{||v_1||^2} * ||v_1||^2$
>
> $<v_1,w_2> = <v_1,v_2> - <v_1,v_2>$
>
> $<v_1,w_2> = 0_{\mathbb{R}^2}$

## Projection of vector(s) to U: $P_u(.)$

$P_u(.)$ is surjective but not injective:

> $N_v = dim(ker(f)) + rk(f)$
>
> $N_v = dim(ker(P_u(.))) + rk(P_u(.))$
>
> $N_v = dim(ker(P_u(.))) + R_u$
>
> $dim(ker(P_u(.))) > 0$ since $N_v > R_u$

__def__ 

$U^\perp = \{v \in V | <v,u> = 0, \forall u \in U\}$

$U \subset V$ and $V = U \oplus U^\perp$ which is different from saying $ V = U \cup U^\perp$

> $V = U \oplus U^\perp$ implies
> 
> $\forall v \in V \rightarrow v = u + u^\perp$, with $u \in U$ and $u^\perp \in U^\perp$

$dim(V) = dim(U) + dim(U^\perp)$

## Isometries

__def:__ linear map $f: V \rightarrow V$, an endomorphism.

f is an isometry if $<v, w> = <f(v), f(w)>$ 

> allows rotation, reflection/symmetry, translation
>
> preserves angles, distances and pointing at the origin

$||v||^2 = <v,v> = <f(v),f(v)> = ||f(v)||^2 \rightarrow ||v-w||^2 = ||f(v) - f(w)||^2$

The matrix associated to f is invertible iff $ker(f) = O_V$
