## Legal bases for processing personal data

1. **Consent**
2. **Contract necessity**: for the performance of a contract between the controller and subject
3. **Compliance with a Legal Obligation**
4. **Task in the Public Interest**: official functions or a task in the public interest
5. **Protecting Vital Interests of the data subject**, e.g., to protect someone's life during medical emergency (life and death)
6. **Legitimate interest of the data controller**: balanced against the rights and freedoms of the individual

<u>example:</u> advertising defends that they have legitimate interest in people's data and that people are interested in advertisement. 

**Freely-given consent**: not valid when there is no real choice 
- **imbalance of power**: presumption of imbalance of power if controller is public authority, employer, medical service. Data subject fearing adverse consequences has no realistic alternative to accept the processing terms (invalid consent -- GDPR Recital\$43)
- **conditionality**: (e.g. non-consent implies denial of services)
- **detriment**

## Specific Consent

**Consent must be specific for each purpose**.

<u>Criteria:</u>
- **Purpose specification**: as a safeguard to avoid "function creep" or widening or blurring of purpose, explain what and why
- **granuarlity in consent requests**: separate opt-in for each purpose
- **clear separation of information** on obtaining consent for data processign activities, from information about other matters

**Examples of non-specific, general purposes:** improving users' experience, marketing purposes, IT-security purposes, future research
