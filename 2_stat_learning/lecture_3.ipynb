### PAC Learnable: We want an algorithm $A$ such that

$$h = A(S)$$
$$R_D(h) \le \epsilon$$
with probability lower than $1-\delta$.

PAC learnability implies that there is a target model $h^*\in H, R_D(h^*)=0$. We speak about **realizability**.

### Agnostic PAC Learnable: We no longer have the knowledge or expectation of $h^*$.

**We abandon the idea that the best risk will be zero**

$$R_D(h) \le \underset{h'\in H}{min} R_D(h') + \epsilon$$
with probability lower than $1-\delta$.

### Linear Regression

Dataset $n_1, ..., x_n$ IID Random Variables linked to target variables $y_1, ..., y_n$.

$$H = \{h: x \rightarrow w^Tx+b, w\in \mathbb{R}^d, b\in\mathbb{R}\}$$

Linear regression is about finding a good $h$. An appropriate loss could be the squared error:

$$l(y, y') = (y-y')^2$$

**Empirical Risk Minimization**: $\hat{R}_S(h) = \frac{1}{n}\sum l(y_i, h(x_i)) = \frac{1}{n}\sum(y_i = h(y_i))^2$

$h \in \underset{w\in\mathbb{R}^d, b\in\mathbb{R}}{argmin} \frac{1}{n}\sum(y_i - w^Tx_i - b)^2$
