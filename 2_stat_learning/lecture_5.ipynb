Let H be a finite class of hypotheses, then clearly $|H_c| \le |H|$. The number of functions from C to {0,1} = $2^{|c|}$.

C cannot be shattered if $|H| \lt 2^{|C|}$.

**Consequence:** $VC(H)\le log_2|H|$ (we have already encountered this logarithm.

**Remark:** The VC dimension can be much smaller.

**Example:** Threshold functions with thresholds in {1, ..., k}

### VC Dimension and number of parameters

The VC dimension coincides with the **number of parameters used to describe the class**.

e.g.

- threshold functions: $VC(H) = 1$
- axis-aligned rectangles: $VC(H) = 2d$

*This is not always the case*

One can even find classes described by one (1) parameter such that $VC(H) = \infty$

Agnostic PAC learnable:

$$m_H(\epsilon, \delta) \approx \frac{d+log(1/\delta)}{\epsilon^2}$$

if $n \ge \frac{d}{\epsilon^2}$ with $\delta = 0.01$

$$R_D(h_s) \le \underset{h \in H}{min}R_D(h) + \epsilon$$

**estimation error** $\le\sqrt{\frac{VCdim(H)}{n}}$ (upper bound)
